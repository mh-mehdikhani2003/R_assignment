```r
# خواندن داده‌ها
library(readr)
data <- read_csv("data.csv")

# دسته‌بندی متغیرهای X و Y
data$X <- cut(
  data$record_max_temp,
  breaks = c(-Inf, 70, 100, Inf),
  labels = c("X3", "X2", "X1"),
  right = TRUE
)

data$Y <- cut(
  data$record_min_temp,
  breaks = c(-Inf, 20, 40, Inf),
  labels = c("Y3", "Y2", "Y1"),
  right = TRUE
)

# محاسبه توزیع مشترک
table_XY <- table(data$X, data$Y)
prob_matrix <- prop.table(table_XY)  # تبدیل به احتمال

# نمایش ماتریس احتمال
print(prob_matrix)
```

        <font face="XB Zar" size=5>
            ابتدا 10000 لیست 15 تایی(مرتب) از متغیرهای تصادفی iid با توزیع
            Uniform(0,1)
            تولید کنید. سپس عناصر اول مربوط به هر لیست را در یک بردار ذخیره کنید(مثلا a).
            همینکار را برای عضو های 10ام و 15ام انجام دهید و در 2 بردار ذخیره کنید(مثلا b و c).
            حال به ازای هر بردار، از توزیع بتای متناسب(با انتخاب پارامتر های مناسب) با آن به اندازه طول بردار نمونه گیری کنید و نمودار چگالی مربوط   
            به بردار و نمونه ها را با هم رسم کنید.
            سپس مقدار اکسپکتد تمام 15 مقدار را محاسبه کنید. و نتیجه را گزارش کنید.           
            از سرچ کردن واژه orderstats در اینترنت هم غافل نشوید.
        </font>
        </div>
    </font>
</div>

```r

total_samples <- 10000
list_size <- 15
samples <- replicate(total_samples, sort(runif(list_size)))

a <- samples[1, ]   # عناصر اول هر لیست
b <- samples[10, ]  # عناصر دهم هر لیست
c <- samples[15, ]  # عناصر پانزدهم هر لیست

# نمونه‌گیری از توزیع بتا برای بردارها
set.seed(42)
a_beta <- rbeta(length(a), 1, list_size)
b_beta <- rbeta(length(b), 10, list_size - 10 + 1)
c_beta <- rbeta(length(c), 15, list_size - 15 + 1)

# رسم نمودارهای چگالی
density_plot <- function(original, sampled, title) {
  plot(density(original), col="blue", lwd=2, main=title, xlab="Value")
  lines(density(sampled), col="red", lwd=2, lty=2)
  legend("topright", legend=c("Original", "Beta Sampled"), col=c("blue", "red"), lwd=2, lty=c(1,2))
}

density_plot(a, a_beta, "Density of a vs Beta Sample")
density_plot(b, b_beta, "Density of b vs Beta Sample")
density_plot(c, c_beta, "Density of c vs Beta Sample")

# محاسبه مقدار مورد انتظار برای تمام 15 مقدار
expected_values <- sapply(1:list_size, function(k) k / (list_size + 1))
print(expected_values)
```
